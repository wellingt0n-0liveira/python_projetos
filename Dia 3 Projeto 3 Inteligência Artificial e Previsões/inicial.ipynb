```python
# Analisando um cliente
# se o cliente é bom, ok ou ruim (Notyas de credito dele = score de credito)

# Passo a passo do projeto
# Passo 0: entender o desafio da empresa
# Passo 1: Importar base de dados
# pip install pandas scikit-learn matplotlib seaborn
%pip install pandas scikit-learn matplotlib seaborn


import pandas as pd

tabela = pd.read_csv("clientes.csv")
display(tabela)
display(tabela.info())

# Passo 2:Tratar e preparar a base dados para a inteigencia artificial
# Passo 3: Criar modelo de previsão, modelo de IA  (machine learning) ruim, ok, Boa
# Passo 4: Avaliar e escolher o melhor modelo
# Passo 5: Fazer novas previsões
```

```python
# Passo 2:Tratar e preparar a base dados para a inteigencia artificial
import pandas as pd

from sklearn.preprocessing import LabelEncoder

# profissao - cientista, marcineiro, pedreiro, pintor, medico

#LabelEncoder

#cientista = 1
#marceneiro = 2
#pedreiro = 3
#pintor = 4
#medico = 5
codificador_profissao = LabelEncoder()
tabela["profissao"] = codificador_profissao.fit_transform(tabela["profissao"])

# mix de credito
codificador_mix = LabelEncoder()
tabela["mix_credito"] = codificador_mix.fit_transform(tabela["mix_credito"]) 

# comportamento pagamento
codificador_comportamento = LabelEncoder()
tabela["comportamento_pagamento"] = codificador_comportamento.fit_transform(tabela["comportamento_pagamento"]) 


# quero prever a coluna scorfe de credito
#display(tabela.info())
display(tabela)


```

```python
# como fuunciona o processo de criar uma IA

# y é o que eu quero prever (score de credito)
y = tabela["score_credito"] 

# x é quem eu vo usar para fazer a previsão
x = tabela.drop(columns=["score_credito", "id_cliente"])

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=42)

# prever o score de credito 
```

```python
# Passo 3: Criar modelo de previsão, modelo de IA  (machine learning) ruim, ok, Bom

# arvore de decisão -> vizinhos proximos
# KNN -> nearest neighbors

# Paso 1:importar a IA

from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

# Passo 2: criar a IA
modelo_arvorededecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()


# treinar a IA (modelo)
modelo_arvorededecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)



```

```python
# Passo 4: Avaliar e escolher o melhor modelo

# previsão
previsao_arvorededecisao = modelo_arvorededecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)



#comparar a previsão com o y_teste
from sklearn.metrics import accuracy_score

print(accuracy_score(y_teste, previsao_arvorededecisao))
print(accuracy_score(y_teste, previsao_knn))

# como melhorar os resultados do modelo.
# GridSearch
# ter mais dados
# testar outros parametros nos modelos
# testar outros modelos
```

```python
# Passo 5: Fazer novas previsões

# melhor modelo é o arvore de decisao

# imprtar os novos clientes
novos_clientes = pd.read_csv("novos_clientes.csv")


novos_clientes["profissao"] = codificador_profissao.transform(novos_clientes["profissao"])
novos_clientes["mix_credito"] = codificador_mix.transform(novos_clientes["mix_credito"]) 
novos_clientes["comportamento_pagamento"] = codificador_comportamento.transform(novos_clientes["comportamento_pagamento"]) 

previsao = modelo_arvorededecisao.predict(novos_clientes)

display(novos_clientes)
display(previsao)

# como adicionar o resultado da previsão na tabela de novos clientes como um acoluna chamada previsao
#novos_clientes["previsao"] = previsao



```
